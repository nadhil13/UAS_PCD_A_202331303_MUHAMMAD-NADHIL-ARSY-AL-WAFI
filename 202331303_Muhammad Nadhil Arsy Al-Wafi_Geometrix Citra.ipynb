## IMPORT LIBRARY

```python
# 202331303_Muhammad Nadhil Arsy Al-Wafi
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## MEMBACA CITRA

```python
# 202331303_Muhammad Nadhil Arsy Al-Wafi
# Membaca gambar
img1 = cv2.imread('geometrik.jpg')
img2 = cv2.imread('geometrik bukti.jpg')

# Mengonversi warna BGR ke RGB
img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

# Membuat subplot
fig, axs = plt.subplots(1, 2, figsize=(12, 8))

# Menampilkan gambar pertama
axs[0].imshow(img1_rgb)
axs[0].axis('off')

# Menampilkan gambar kedua
axs[1].imshow(img2_rgb)
axs[1].axis('off')

plt.show()
```

## CITRA ASLI

```python
# 202331303_Muhammad Nadhil Arsy Al-Wafi
original = img1.copy()
```

## TRANSFORMASI

```python
# 202331303_Muhammad Nadhil Arsy Al-Wafi
(h, w) = img1.shape[:2]
center = (w // 2, h // 2)
M_rotate = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(img1, M_rotate, (w, h))
resized = cv2.resize(img1, (60, 120))
cropped = img1[100:400, 150:350]
flipped = cv2.flip(img1, 1)
M_translate = np.float32([[1, 0, 50], [0, 1, 30]])
translated = cv2.warpAffine(img1, M_translate, (w, h))
```

## HASIL

```python
# 202331303_Muhammad Nadhil Arsy Al-Wafi
fig, axs = plt.subplots(2, 3, figsize=(12, 8))

axs[0, 0].imshow(original)
axs[0, 0].set_title("Citra Asli")

axs[0, 1].imshow(rotated)
axs[0, 1].set_title("After Rotated")

axs[0, 2].imshow(resized)
axs[0, 2].set_title("After Resized")

axs[1, 0].imshow(cropped)
axs[1, 0].set_title("After Cropped")

axs[1, 1].imshow(flipped)
axs[1, 1].set_title("After Flipped")

axs[1, 2].imshow(translated)
axs[1, 2].set_title("After Translated")

plt.tight_layout()
plt.show()
```
