## IMPORT LIBRARY

```python
# 202331303_Muhammad Nadhil Arsy Al-Wafi
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## MEMBACA GAMBAR

```python
# 202331303_Muhammad Nadhil Arsy Al-Wafi
# Membaca gambar
img1 = cv2.imread('filtering citra.jpg')
img2 = cv2.imread('filtering citra bukti.jpg')
gray = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)

# Mengonversi warna BGR ke RGB
img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

# Membuat subplot
fig, axs = plt.subplots(1, 2, figsize=(12, 8))

# Menampilkan gambar pertama
axs[0].imshow(img1_rgb)
axs[0].axis('off')

# Menampilkan gambar kedua
axs[1].imshow(img2_rgb)
axs[1].axis('off')

plt.show()
```

## MEDIAN FILTERING (KITA AKAN GUNAKAN BGR AGAR TETAP BERWARNA OBJEK)

```python
# 202331303_Muhammad Nadhil Arsy Al-Wafi
median_filtered = cv2.medianBlur(img1, 3)
```

## MEAN FILTERING MANUAL PADA GRAYSCALENYA

```python
# 202331303_Muhammad Nadhil Arsy Al-Wafi
def manual_mean_filter(image, kernel_size=3):
    pad = kernel_size // 2
    padded = np.pad(image, pad, mode='edge')
    filtered = np.zeros_like(image)
    
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            region = padded[i:i+kernel_size, j:j+kernel_size]
            filtered[i, j] = int(np.mean(region))
    return filtered

mean_filtered = manual_mean_filter(gray)
```

## Menampilkan

```python
# 202331303_Muhammad Nadhil Arsy Al-Wafi
fig, axs = plt.subplots(2, 2, figsize=(12, 12))

# Baris 1
axs[0, 0].imshow(gray, cmap='gray')
axs[0, 0].set_title('Citra Asli')

axs[0, 1].imshow(cv2.cvtColor(median_filtered, cv2.COLOR_BGR2RGB))
axs[0, 1].set_title('After filter median')

# Baris 2
axs[1, 0].imshow(gray, cmap='gray')
axs[1, 0].set_title('Original Image')

axs[1, 1].imshow(mean_filtered, cmap='gray')
axs[1, 1].set_title('Mean Filtered Image')

plt.tight_layout()
plt.show()
```
